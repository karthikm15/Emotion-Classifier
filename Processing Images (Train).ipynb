```python
from PIL import Image, ImageEnhance
from pathlib import Path
import matplotlib.pyplot as plt
import uuid
import random
import os
import shutil
```

```python
def sharpen(img, factor):
    enhancer_sharpness = ImageEnhance.Sharpness(img)
    return enhancer_sharpness.enhance(factor)

def contrast(img, factor):
    enhancer_contrast = ImageEnhance.Contrast(img)
    return enhancer_contrast.enhance(factor)

def color(img, factor):
    enhancer_color = ImageEnhance.Color(img)
    return enhancer_color.enhance(factor)

def rotate(img, degrees):
    return img.rotate(degrees)
```

```python
randContrastMin, randContrastMax = (0.8, 1.2)
randSharpenMin, randSharpenMax  = (0.8, 1.2)
randColorMin, randColorMax    = (0.8, 1.2)
multiplier = 2
```

```python
inputBasePath_t = './train/'
outputBasePath_t  = './train_processed/'
folders = ['anger', 'contempt', 'disgust', 'fear', 'sadness', 'surprise']
max_folders = ['neutral', 'happiness']
```

```python
inputBasePath_v = './valid/'
outputBasePath_v  = './valid_processed/'
```

```python
unique_emotions = ['anger', 'surprise', 'disgust', 'fear', 'neutral', 'happiness', 'sadness', 'contempt']
multiplier_emotions = []
for emotions in unique_emotions:
    counter = 0
    for filename in os.listdir("./train/" + emotions):
        counter += 1
#         if (counter % 100 == 
    print (emotions + " " + str(counter))
    multiplier = round(4000/counter)
    multiplier_emotions.append(multiplier)
    
print(multiplier_emotions)
```

```python
index = 0
for f in folders:
    plist = Path(inputBasePath_t + f + '/').glob('*.jpg')

    outpath = outputBasePath_t + f + '/' 
    if not os.path.exists(outpath):
        os.makedirs(outpath)
    
    for path in os.listdir("./train/" + f):
        print(path)
        
        if (path != ".ipynb_checkpoints"):
            i = Image.open("./train/" + f + "/" + path)

            for m in range(multiplier_emotions[unique_emotions.index(f)]):

                randContrast = random.uniform(randContrastMin, randContrastMax)
                randSharpen = random.uniform(randSharpenMin, randSharpenMax)
                randColor = random.uniform(randColorMin, randColorMax)
                i = contrast(i, randContrast)
                i = sharpen(i, randSharpen)
                i = color(i, randColor)

                i.save(outputBasePath_t + f + '/' + str(m) + path) 
                print('.', end='')
    index += 1
```

```python
count = 0
for f in max_folders:
    
    outpath = outputBasePath_t + f + '/' 
    if not os.path.exists(outpath):
        os.makedirs(outpath)
        
    for path in os.listdir("./train/" + f):
        source = './train/' + f + '/' + path
        destination = outputBasePath_t + f + '/' + path
        
        if (not os.path.exists(destination)):
            shutil.copy(source,destination)
            
        print(source + " " + str(count))
        
        count += 1
```
