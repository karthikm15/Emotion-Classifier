```python
import PIL
from PIL import Image
import os
import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import boto3
import shutil
```

```python
# !mkdir train_resize
# !mkdir valid_resize
unique_emotions = ['anger', 'surprise', 'disgust', 'fear', 'neutral', 'happiness', 'sadness', 'contempt']
```

```python
def prepare(filepath, IMG_SIZE):
    img_array = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE) 
    new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE)) 
    return new_array.reshape(-1, IMG_SIZE, IMG_SIZE)
```

```python
unique_emotions = ['anger', 'surprise', 'disgust', 'fear', 'neutral', 'happiness', 'sadness', 'contempt']
index = 0
for emotions in unique_emotions:
    for filename in os.listdir("./train_processed/" + emotions):
        destination = './train_resize/'+ emotions + "/" + filename
        
        print(filename + " " + emotions)
        
        if (filename == ".ipynb_checkpoints"):
            continue
        else:
            image_data = prepare(('./train_processed/' + emotions + "/" + filename) , 50)
            image = Image.fromarray(image_data[0])
            try:
                image.save('./train_resize/' + emotions + "/" + filename)
            except:
                os.mkdir('train_resize/'+emotions)
                image.save('./train_resize/' + emotions + "/" + filename)
```
