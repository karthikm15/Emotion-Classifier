```python
import pandas as pd
import os
import shutil
import os.path
import numpy as np
```

```python
df = pd.read_csv("facial_expressions/data/legend.csv")
del df['Unnamed: 0']
del df['user.id']
df.head()
```

```python
len(df)
```

```python
unique_emotions = list(df['emotion'].unique())
unique_emotions
```

```python
df = df.sample(frac = 1) 
train = df[0:int(len(df['image'])*0.6)]
valid = df[int(len(df['image'])*0.6):]
train = train.reset_index()
valid = valid.reset_index()
```

```python
!mkdir train
!mkdir valid
```

```python
for index in train.index:
    if (index % 100 == 0):
        print(index)
    source = "./facial_expressions/images/"+train['image'][index]
    destination = './train/'+train['emotion'][index]+'/'+train['image'][index]
    if (os.path.exists(source) and not (os.path.exists(destination))):
        try:
            dest = shutil.move(source, destination)
        except:
            os.mkdir('train/'+train['emotion'][index])
            dest = shutil.move(source,destination)
```

```python
for index in valid.index:
    if (index % 100 == 0):
        print(index)
    source = "./facial_expressions/images/"+valid['image'][index]
    destination = './valid/'+valid['emotion'][index]+'/'+valid['image'][index]
    
    if (os.path.exists(source) and not (os.path.exists(destination))):
        try:
            dest = shutil.move(source, destination)
        except:
            os.mkdir('valid/'+valid['emotion'][index])
            dest = shutil.move(source,destination)
```

```python
os.remove("./output/valid_lst.lst")
```

```python
file = open('./output/train_lst.lst', 'w+')
emotions = unique_emotions
counter = 0
for index in train.index:
    val_index = unique_emotions.index(train['emotion'][index])
    source = "./train/"+unique_emotions[val_index]+"/"+train['image'][index]
    output_path = './output/'+unique_emotions[val_index]+"/"+train['image'][index]
    
    path_in_file = unique_emotions[val_index]+"/"+train['image'][index]
    
    
    if (os.path.exists(source) and not (os.path.exists(output_path))):
        try:
            dest = shutil.move(source, output_path)
        except:
            os.mkdir('output/'+train['emotion'][index])
            dest = shutil.move(source,output_path)
            
    if (os.path.exists(output_path)):
        file.write('%d\t%d\t%s\n' % (counter, val_index, path_in_file))
        counter += 1
```

```python
file = open('./output/valid_lst.lst', 'w+')
emotions = unique_emotions
counter = 0
for index in valid.index:
    val_index = unique_emotions.index(valid['emotion'][index])
    source = "./valid/"+unique_emotions[val_index]+"/"+valid['image'][index]
    output_path = './output/'+unique_emotions[val_index]+"/"+valid['image'][index]
    
    path_in_file = unique_emotions[val_index]+"/"+valid['image'][index]
    
    if (os.path.exists(source) and not (os.path.exists(output_path))):
        try:
            dest = shutil.move(source, output_path)
        except:
            os.mkdir('output/'+valid['emotion'][index])
            dest = shutil.move(source,output_path)
    
    if (os.path.exists(output_path)):
        file.write('%d\t%d\t%s\n' % (counter, val_index, path_in_file))
        counter += 1
```
